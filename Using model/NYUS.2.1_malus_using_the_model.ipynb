# NYUS.2.1.malus using the model for prediction

The NYUS.2.1.malus model is a newer version of the NYUS.2.1 model that could use to predict the cold hardiness of grapevine and apple.

## Using NYUS.2.1.malus
Before using NYUS.2.1.malus, Please generate a dataframe that contains all the features using this script: https://github.com/imbaterry11/NYUS.2.1_apple_grape/blob/main/Raw%20data%20processing%20and%20feature%20extraction/Feature_extraction.R
The NYUS.2.1.malus model folder should be under the same dir of this notebook. The NYUS.2.1.malus model can be downloaded here: 
https://cornell.box.com/s/6t9bol0q8ejwhwynhqsxxzrk55voxsog

```python
import autogluon
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor_LT50 = TabularPredictor.load("NYUS2_1_malus",require_version_match=False) 
```

```python
df=pd.read_csv('daily_temperature_data_example_feature_extracted.csv', sep=",", header=0)
```

```python
df
```

```python
df_pred = df.drop(['Date'],axis = 1)
```

```python
#Prediction
y_pred = predictor_LT50.predict(df_pred)
```

```python
#Save prediction result:
y_pred.to_csv (r'LT50_pred.csv', index = True, header=True)
```
